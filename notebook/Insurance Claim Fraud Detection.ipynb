##### custom mapping for encoding

```python
cat_df['insured_education_level'] = cat_df['insured_education_level'].map({'High School':1, 'College':2, 'PhD':3, 'Masters':4})
cat_df['police_report_available'] = cat_df['police_report_available'].map({'Yes':1, 'No':0})
cat_df['incident_severity'] = cat_df['incident_severity'].map({'Total Loss':1, 'Minor Damage':2, 'Major Damage':3})
cat_df['fraud_reported'] = cat_df['fraud_reported'].map({'N':0,'Y':1})
cat_df['insured_sex'] = cat_df['insured_sex'].map({'OTHER':2, 'MALE':1, 'FEMALE':0})
cat_df['insured_occupation'] = cat_df['insured_occupation'].map
```

```python
cat_df.columns
```

```python
for col in cat_df.drop(columns=['insured_education_level','police_report_available','incident_severity','fraud_reported','insured_sex']).columns:
    cat_df= pd.get_dummies(cat_df,columns=[col], prefix=[col],drop_first=True)
```

```python
label_cols = [
    'insured_sex',
    'insured_education_level',
    'incident_severity',
    'police_report_available',
    'fraud_reported'
]
```

```python
onehot_cols = [col for col in cat_df.columns if col not in label_cols]
```

```python
cat_df = pd.get_dummies(
    cat_df,
    columns=onehot_cols,
    drop_first=True,
    dtype=int
)
```

#### KNeighborsClassifier

```python
from sklearn.neighbors import KNeighborsClassifier

knn_model = KNeighborsClassifier(n_neighbors = 30)
knn_model.fit(x_train, y_train)

y_pred = knn_model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

knn_model_train_acc = accuracy_score(y_train, knn_model.predict(x_train))
knn_model_test_acc = accuracy_score(y_test, y_pred)

print("Training Accuracy: ", knn_model_train_acc)
print("Testing Accuracy: ", knn_model_test_acc)
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

#### Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
decision_tree_model = DecisionTreeClassifier()
decision_tree_model.fit(x_train,y_train)
y_pred = decision_tree_model.predict(x_train)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid_params = {
    'criterion' : ['gini', 'entropy'],
    'max_depth' : [3, 5, 7, 10],
    'min_samples_split' : range(2, 10, 1),
    'min_samples_leaf' : range(2, 10, 1)
}

grid_search = GridSearchCV(decision_tree_model,grid_params, cv = 5, n_jobs = -1,verbose= 1)
grid_search.fit(x_train,y_train)
```

```python
print(grid_search.best_params_)
print(grid_search.best_score_)
```

```python
decision_tree_model = grid_search.best_estimator_
y_pred = decision_tree_model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
decision_tree_model_train_acc = accuracy_score(
    y_train,
    decision_tree_model.predict(x_train)
)

decision_tree_model_test_acc = accuracy_score(
    y_test,
    y_pred
)
print("Training Accuracy:", decision_tree_model_train_acc)
print("Testing Accuracy:", decision_tree_model_test_acc)
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
